### Lets do univariate analysis on each variable

```python
def Univariate_Analysis(df,column):
        sns.set(style="whitegrid")
        fig, ax=plt.subplots(nrows =1,ncols=2,figsize=(18,6))
        ax[0].set_title("Distribution Plot")
        sns.distplot(df[column],ax=ax[0])
        ax[1].set_title("Box Plot")
        sns.boxplot(data =df, y=column,ax=ax[1],orient='v')
        plt.show()
```

```python
Univariate_Analysis(df=BS,column='Count')
# it can be seen that data is normally distributed and mean is between 4000 and 6000
```

```python
Univariate_Analysis(df=BS,column='windspeed')
# distribution chart seems normally distributed, little bit tilted on left side and mean is between 10 and 15
```

```python
Univariate_Analysis(df=BS,column='temp')
# distribution chart seems normally distributed and mean is between 20-22
```

# Data preperation

### Lets start making dummy variables


```python
Months1=pd.get_dummies(BS.Month,drop_first=True)
weekdays1=pd.get_dummies(BS.weekday,drop_first=True)
weathersit1=pd.get_dummies(BS.weathersit,drop_first=True)
season1=pd.get_dummies(BS.season,drop_first=True)
BS=pd.concat([Months1,weekdays1,weathersit1,season1,BS],axis=1)
BS.head()
```

```python
# lets drop following columns
BS.drop(['season','Month','weekday','weathersit'], axis = 1, inplace = True)
BS.head()
```

## Analysis on the given data
#### - As we can see, the temperature variable has the largest coefficient, 0.5122, which means that for every unit increase in temperature, there is a corresponding increase in the number of bicycle rentals of 0.5122 units.

#### Furthermore, we observe that several variables have negative coefficients. A negative coefficient indicates that the dependent variable tends to drop as the independent variable rises.

#### Spring, hazy mist, and light snow are variables with negative coefficients. When other model variables are held constant, the coefficient value represents how much the dependent variable changes when the independent variable is moved by one unit.
